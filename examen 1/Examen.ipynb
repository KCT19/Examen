```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

tamaño_tablero = 7
tamaño_casilla = 100

colores = [
    [255, 255, 255]
    [255, 255, 0]
    [255, 102, 0]
    [255, 0, 0]            
    [153, 0, 153]
    [0, 51, 255]        
    [0, 153, 0]    
]

tablero = np.zeros((tamaño_tablero, tamaño_tablero, 3), dtype=np.uint8)

for i in range(tamaño_tablero):
    for j in range(tamaño_tablero):
        tablero[i, j] = colores[(i + j) % len(colores)]

tablero_ampliado = np.kron(tablero, np.ones((tamaño_casilla, tamaño_casilla, 1), dtype=np.uint8))


plt.figure(figsize=(7, 7))
plt.imshow(tablero_ampliado)
plt.xticks([])
plt.yticks([])
plt.title("Ejercicio2")
plt.show()


```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
imagen = cv2.imread('Imagen1.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
alto, ancho, _ = imagen.shape
degradado = np.linspace(0, 1, alto, dtype=np.float32).reshape(alto, 1)
degradado = np.dstack([degradado] * 3) 
imagen_degradada = (imagen * degradado).astype(np.uint8)
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen)
plt.title('Imagen Original')

plt.subplot(1, 2, 2)
plt.imshow(imagen_degradada)
plt.title('Imagen con Degradado Lineal a Colores')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('Imagen1.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

alto, ancho, _ = imagen.shape

imagen_negra = np.zeros((alto, ancho, 3), dtype=np.uint8)
desplazamiento_y = -75

puntos_triangulo = np.array([
    [ancho // 2, alto // 4 + desplazamiento_y],     
    [ancho // 4, 3 * alto // 4 + desplazamiento_y],   
    [3 * ancho // 4, 3 * alto // 4 + desplazamiento_y] 
], np.int32)

mask = np.zeros((alto, ancho), dtype=np.uint8)
cv2.fillPoly(mask, [puntos_triangulo], 255)

imagen_final = np.zeros_like(imagen)
imagen_final[mask == 255] = imagen[mask == 255]

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(imagen)
plt.title('Imagen Original')


plt.subplot(1, 2, 2)
plt.imshow(imagen_final)
plt.title('Triangulo')


plt.tight_layout()
plt.show()


```

Teoría 

1.- ¿A qué me refiero si hablo de una imagen digital?
Cuando hablo de una imagen digital, me estoy refiriendo a una representación visual de un objeto o escena que ha sido convertida en datos numéricos. Esta representación se compone de una matriz de píxeles, donde cada uno de estos píxeles tiene un valor asociado que representa su color o intensidad. Las imágenes digitales son las que veo en pantallas de computadoras, teléfonos o cámaras, y se pueden almacenar, editar y procesar en dispositivos electrónicos.

2. ¿Qué es la profundidad del color?
La profundidad del color hace referencia a cuántos bits se utilizan para representar el color de cada píxel en una imagen. Cuanto mayor es la profundidad del color, más colores pueden representarse. Por ejemplo, con 8 bits por canal, una imagen puede tener hasta 256 colores diferentes por canal, lo que genera 16,7 millones de colores en total (cuando se usan 3 canales, como en el modelo RGB). La profundidad del color influye en la calidad y la cantidad de detalles visuales que puedo ver en una imagen.

3. ¿Cuál es la diferencia entre CMYK y RGB?
La diferencia entre CMYK y RGB radica en cómo se crean los colores:

RGB (Rojo, Verde, Azul) es un modelo de color aditivo. En este modelo, los colores se crean sumando luz. Cuanto más se combinan los colores rojo, verde y azul, más se acercan al blanco. Este modelo lo uso para ver imágenes en pantallas, como en computadoras o teléfonos, porque esas pantallas emiten luz.

CMYK (Cian, Magenta, Amarillo, Negro) es un modelo de color sustractivo. Aquí, los colores se crean restando luz a partir de un fondo blanco, que es el color natural del papel. Utilizo este modelo cuando imprimo algo, porque las impresoras aplican tintas de estos colores para formar el resultado final.
